```python
import csv
import pandas
import numpy
import matplotlib.pyplot as plt 
import scipy
import math
```

```python
df = pandas.read_csv('NIFTY50.csv')
```

```python
closingPoints = df.Close
openingPoints = df.Open
droplist = []
for i in range(len(closingPoints)):
    droplist.append(closingPoints[i] - openingPoints[i])

maxdrop = max(droplist)
mindrop = -min(droplist)
```

```python
plt.plot(droplist, label = "Drop in a day") 


plt.xlabel('Time') 
plt.ylabel('Points') 

plt.title('NIFTY50 since its inception') 

plt.legend() 

plt.show() 

print("The Maximum drop that the NIFTY50 index noticed since its inception is " + str(maxdrop) + "points.\nThis drop was witnessed on " + str(df.Date[droplist.index(maxdrop)]) + ".")
print("The Maximum rise that the NIFTY50 index noticed since its inception is " + str(mindrop) + "points.\nThis drop was witnessed on " + str(df.Date[droplist.index(mindrop)]) + ".")
```
